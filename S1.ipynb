```java
%maven com.google.guava:guava:33.2.1-jre
%maven tech.tablesaw:tablesaw-core:0.44.4
import tech.tablesaw.api.*;
import tech.tablesaw.columns.*;
import tech.tablesaw.io.csv.CsvReadOptions;

CsvReadOptions options = CsvReadOptions.builder("Book.csv")
        .header(false)
        .build();
Table booksTable = Table.read().usingOptions(options);

booksTable.column("C0").setName("ISBN");
booksTable.column("C1").setName("title");
booksTable.column("C2").setName("edition");
booksTable.column("C3").setName("year");

booksTable.stringColumn("title").print();
```
